```python
import pandas as pd
```

```python
df=pd.read_csv("heart.csv")
df
```

```python
df["count"]=1
df
```

```python
df_female=df[df['sex']==0]
df_female
```

```python
df_female_age=df_female.groupby('age_range').agg('sum').reset_index()
```

```python
df_female_age['Proportion of Heart Attack']=df_female_age['output']/df_female_age['count']
df_female_age
```
